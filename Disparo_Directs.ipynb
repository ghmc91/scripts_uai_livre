```python
import pandas as pd
import time

df = pd.read_excel('C:/Users/Usuario/Documents/UaiLivre/arquivos/Sellers.xlsx')
df['Produção Mensal'] = df['Produção Mensal'].astype(int)
df = df.loc[df['Produção Mensal'].between(9,21)]
df = df.dropna(subset=['instagram'])
df['instagram'] = df['instagram'].str.split()
df['instagram'] = [x[0] for x in df['instagram']]
df['instagram'] = df.instagram.str.strip('@')
df = df.drop_duplicates(subset=['instagram'])
```

```python
index_xakriaba = df.loc[df['instagram'].isin(['ceramica_indigena_xakriaba', 'monicafonsecar'])].index
df.drop(index_xakriaba, inplace=True)

df = df[7:]
insta_list = df['instagram'].to_list()
```

```python
import requests

lista_final = []
for i in insta_list:
    response = requests.get("https://instagram.com/" + i + "/")
    if response.status_code != 404 :
        lista_final.append(i)
```
